```python
import pandas as pd
import numpy as np

np.random.seed(42)

n = 500

hours_studied = np.random.randint(1, 10, n)
attendance = np.random.randint(60, 100, n)
sleep_hours = np.random.randint(4, 9, n)
assignments_completed = np.random.randint(1, 10, n)

score = (
    hours_studied*5
    + attendance*0.4
    + sleep_hours*2
    + assignments_completed*3
    + np.random.normal(0,5,n)
)

score=np.clip(score,0,100)

df=pd.DataFrame({
    'Hours_Studied':hours_studied,
    'Attendance':attendance,
    'Sleep_Hours':sleep_hours,
    'Assignments_Completed':assignments_completed,
    'Score':score
})

df.to_csv("../data/student_scores.csv",index=False)

df.head()
```
